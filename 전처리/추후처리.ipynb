```python
import pandas as pd
import numpy as np

df_train = pd.read_csv("train.csv") # 학습용 데이터
df_test = pd.read_csv("submission0225_3__lgb3개2.csv")
```

### 의미있어보이는 행을 True로

```python
cross_tab = pd.crosstab(df_train['bant_submit'], df_train['is_converted'])

# True 비율 계산
cross_tab['ratio'] = cross_tab[True] / (cross_tab[False] + cross_tab[True])
cross_tab5 = cross_tab.sort_values(by='ratio', ascending=False)
cross_tab5.head(50)
```

```python
high_ratio_response = cross_tab[cross_tab['ratio'] >= 0.01].index.tolist()
```

#### 1-1번째 기준

```python
cross_tab = pd.crosstab(df_train['lead_owner'], df_train['is_converted'])

# True 비율 계산
cross_tab['ratio'] = cross_tab[True] / (cross_tab[False] + cross_tab[True])
cross_tab1 = cross_tab.sort_values(by='ratio', ascending=False)
cross_tab1.head(110)
```

```python
cross_tab = pd.crosstab(df_train['lead_owner'], df_train['is_converted'])
cross_tab['ratio'] = cross_tab[True] / (cross_tab[False] + cross_tab[True])

# 'ratio'가 0.65 이상인 'lead_owner'인것
high_ratio_owners = cross_tab[cross_tab['ratio'] >= 0.05].index.tolist()

```

#### 2번째 기준

```python
cross_tab = pd.crosstab(df_train['customer_idx'], df_train['is_converted'])

# True 비율 계산
cross_tab['ratio'] = cross_tab[True] / (cross_tab[False] + cross_tab[True])
cross_tab3 = cross_tab.sort_values(by='ratio', ascending=False)
cross_tab3.head(1500)
```

```python
high_ratio_customer = cross_tab3[cross_tab3['ratio'] >= 0.01].index.tolist()
```

#### 3번째 기준

```python
df_test.loc[
    (df_test['customer_idx'].isin(high_ratio_customer) | df_test['customer_idx'].isna()) &
    (df_test['lead_owner'].isin(high_ratio_owners) | df_test['lead_owner'].isna()) &
    (df_test['customer_job'].isin(high_ratio_job) | df_test['customer_job'].isna()) &
    (df_test['product_category'].isin(high_ratio_category) | df_test['product_category'].isna()) &
    (df_test['response_corporate'].isin(high_ratio_response) | df_test['response_corporate'].isna()), 
    'is_converted'
] = 1
```

### 의미없어보이는 행을 False로

```python
# False 비율 계산
cross_tab['ratio'] = cross_tab[False] / (cross_tab[False] + cross_tab[True])
cross_tab = cross_tab.sort_values(by='ratio', ascending=False)
cross_tab.head(30)
```

```python
cross_tab = pd.crosstab(df_train['lead_owner'], df_train['is_converted'])
cross_tab['ratio'] = cross_tab[True] / (cross_tab[False] + cross_tab[True])

# 'ratio'가 0.1 이하인 'lead_owner'인것
low_ratio_owners = cross_tab[cross_tab['ratio'] <= 0.01].index.tolist()

```

```python
#그러면서 'ver_pro'와 'ver_cus'가 0인 값
df_test.loc[(df_test['lead_owner'].isin(low_ratio_owners)) & (df_test['ver_pro'] != 1) & (df_test['ver_cus'] != 1) & (df_test['id_strategic_ver'] != 1) & (df_test['it_strategic_ver'] != 1), 'is_converted'] = 0
```

```python
converted_count = df_test['is_converted'].sum()
converted_count
```
