```python
# remove un-used columns
df = df.drop(columns=['product_subcategory', 'product_modelname', 'customer_country.1'])

# null to 0
df.update(df[['historical_existing_cnt','id_strategic_ver','it_strategic_ver', 'idit_strategic_ver']].fillna(0))

# preprocess object-type columns
str_cols = df.columns[df.dtypes==object]
df[str_cols] = df[str_cols].fillna('undefined')  # null to 'undefined'
df[str_cols] = df[str_cols].apply(lambda x: x.str.replace(' ', ''))  # remove spaces
df[str_cols] = df[str_cols].apply(lambda x: x.str.lower())  # lower case

df[str_cols] = df[str_cols].apply(lambda x: x.str.replace(pat=r'[^a-z0-9]', repl=r'', regex=True))  # remove special characters (case 1. all, including non-eng to eng)
# df[str_cols] = df[str_cols].apply(lambda x: x.str.replace(pat=r'[^a-z0-9,/]', repl=r'', regex=True))  # remove special characters (case 2. only a few for tokenizing issue)

# merging
df = df.replace(dict.fromkeys(['etc', 'others'], 'other'))
```
