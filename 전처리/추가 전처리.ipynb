## 이상 컬럼 전처리

### 전체 적용은 df_all로 수정해야함

```python
# customer_country 전처리 뒤에서 80개, count 20개이하는 삭제
category_counts = df_train['customer_country'].value_counts()
categories_to_remove = category_counts[category_counts <= 20].index
df_train.loc[df_train['customer_country'].isin(categories_to_remove), 'customer_country'] = 'undefined'
df_train['customer_country'].unique()
```

```python
## customer_type 전처리
similar_categories_mapping = {
    'other': 'undefined',
    'dealerdistributor': 'distributor',
    'commercialenduser': 'enduser',
}

df_train['customer_type'] = df_train['customer_type'].replace(similar_categories_mapping)
category_counts = df_train['customer_type'].value_counts()
categories_to_remove = category_counts[category_counts <= 5].index
df_train.loc[df_train['customer_type'].isin(categories_to_remove), 'customer_type'] = 'undefined'
```

```python
# customer_job 전처리
similar_categories_mapping = {
    'generalmanager': 'manager',
    'designer': 'artist',
}
df_train['customer_job'] = df_train['customer_job'].replace(similar_categories_mapping)
category_counts = df_train['customer_job'].value_counts()
categories_to_remove = category_counts[category_counts <= 31].index
df_train.loc[df_train['customer_job'].isin(categories_to_remove), 'customer_job'] = 'other'
```

```python
# product_category 전처리
similar_categories_mapping = {
    'other': 'undefined',
    'laptop': 'pc',
}
df_train['product_category'] = df_train['product_category'].replace(similar_categories_mapping)
category_counts = df_train['product_category'].value_counts()
categories_to_remove = category_counts[category_counts <= 60].index
df_train.loc[df_train['product_category'].isin(categories_to_remove), 'product_category'] = 'undefined'
```
